```python
import matplotlib.pyplot as plt
import numpy as np
import math

start = (2, 2)

sweep_direction = -math.pi/4 # -pi/2 to pi/2
line_p_factor = 3000
line_x = np.linspace(start[0] - line_p_factor*math.cos(sweep_direction), start[0] + line_p_factor*math.cos(sweep_direction), line_p_factor)
line_y = np.linspace(start[1] - line_p_factor*math.sin(sweep_direction), start[1] + line_p_factor*math.sin(sweep_direction), line_p_factor)

# line_x = np.linspace(-20, 20, 100)
# line_y = np.linspace(-20, 20, 100)
print(line_x.shape)
# 绘制图表
plt.plot(line_x, line_y, 'grey')

# 设置坐标轴范围
plt.xlim([0, 2])  # X轴范围从0到6
plt.ylim([0, 3]) # Y轴范围从0到3

# 显示图表
plt.show()
```
